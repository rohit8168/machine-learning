```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\ml\da\mnist_train.csv")
l=data['label']
data=data.drop("label",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
data=StandardScaler().fit_transform(data)
```

```python
data_1000=data[0:1000,:]
label_1000=l[0:1000]
```

```python
model=TSNE(n_components=2,random_state=0)
tsne_data=model.fit_transform(data_1000)
tsne_data=pd.DataFrame(tsne_data,columns=["Dim_1","Dim_2"])
tsne_df=pd.concat((tsne_data,label_1000),axis=1)
tsne_df
```

```python
sns.FacetGrid(tsne_df,hue="label").map(plt.scatter,"Dim_1","Dim_2").add_legend()
```
