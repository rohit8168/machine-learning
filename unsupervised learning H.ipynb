```python
from sklearn.datasets import make_blobs
data=make_blobs(500,n_features=5,centers=3)
```

```python
x=data[0]
y=data[1]
```

```python
import pandas  as pd
import numpy as np
x=pd.DataFrame(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[0],x[1],c=x[0])
```

```python
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=3)
y_pred=Kmeans.fit_predict(x,y)
```

<p>centroid intialize method</p>

```python
x=x.iloc[:,0:2]

good_init=np.array([[-7,-9],[5,-4],[-2,5]])
kmeans=KMeans(n_clusters=3,init=good_init,n_init=1)
```

```python
y_pred=kmeans.fit_predict(x,y)
```

```python
good_init[0][0]
```

```python
plt.scatter(x[0],x[1],c=x[0])
plt.scatter(good_init[0][0],good_init[0][1],c="black")
```

```python
kmeans.score(x)
```

```python
from sklearn.cluster import MiniBatchKMeans
clf=MiniBatchKMeans(n_clusters=3)
clf.fit(x)
```

```python
clf.score(x)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,kmeans.labels_)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.05)
db = DBSCAN(eps=0.05, min_samples=5)
db.fit(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=50)
knn.fit(db.components_,db.labels_[db.core_sample_indices_])
```
